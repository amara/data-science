```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

file_name = "my_iris.xlsx"
```

```python
df = pd.read_excel(file_name)

# Dropping Unnamed Column 
# Dropping target column since data is redundant
df.drop([df.columns[0], df.columns[-1]], axis= 1, inplace= True) 
df
```

```python
pca = PCA(n_components= 2)
df_naFill = df.fillna(df.mean())
components = pca.fit_transform(df_naFill)

comp1 = components[:,0]
comp2 = components[:,1]
df_pca = pd.DataFrame([comp1, comp2]).transpose()
Skew = 3 * (df_pca.mean() - df_pca.median()) / df_pca.std()
Skew
print(Skew)

plt.scatter(comp1, comp2)
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

A skewness value greater than 1 or less than -1 indicates a highly skewed distribution. A value between 0.5 and 1 or -0.5 and -1 is moderately skewed. A value between -0.5 and 0.5 indicates that the distribution is fairly symmetrical.

Using Min or Max filling method has a high chance of filling the values with outliers so we wont use that method

Since the data is not highly skewed, we can use the mean for the best results using the average value for filling NA rows.
